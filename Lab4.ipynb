```python

import os
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import dash
from dash import dcc, html, Input, Output
import tkinter as tk
from tkinter import filedialog

# ====================
# Data Loading using Tkinter
# ====================
root = tk.Tk()
root.withdraw()  # Hide the root Tk window
#Caj Rollny; the original code - "filedialog.askopenfilename" - didn't work. Tried adding this line of code after googling, worked:
root.call('wm', 'attributes', '.', '-topmost', True)

print("Imports done, Tk initiated")


# Define file keys for CSV files.
# Caj Rollny: plot only between Mora and Borlänge!
# Tried some datasets "manually" with this code and found the corresponding datasets:
# Mora -> Leksand: 2024-12-10 10-00-00 (1)
# Leksand -> Borlänge: 2024-12-10 12-00-00 (1)
# 1. merge theese datasets into new datasets and save as new files:
# https://stackoverflow.com/questions/48051100/python-pandas-merge-multiple-csv-files

#Caj Rollny: NB! The file generation below took about 15 minutes in total. Uncomment the code below if the files need to be re-generated!
'''
#GPS Latitude:
df_GPS_lat1 = pd.read_csv("C:\Temp\D7015B\Data2\\2024-12-10 10-00-00 (1)\\GPS.latitude.csv", header=None)
df_GPS_lat2 = pd.read_csv("C:\Temp\D7015B\Data2\\2024-12-10 12-00-00 (1)\\GPS.latitude.csv", header=None)
df_GPS_lat_combined = pd.concat([df_GPS_lat1,df_GPS_lat2])
#save combined dataset as new file:
df_GPS_lat_combined.to_csv("C:\Temp\D7015B\GPS.latitude.csv",index=None)
#free memory:
del df_GPS_lat1
del df_GPS_lat2
del df_GPS_lat_combined

#GPS longitude:
df_GPS_lon1 = pd.read_csv("C:\Temp\D7015B\Data2\\2024-12-10 10-00-00 (1)\\GPS.longitude.csv", header=None)
df_GPS_lon2 = pd.read_csv("C:\Temp\D7015B\Data2\\2024-12-10 12-00-00 (1)\\GPS.longitude.csv", header=None)
df_GPS_lon_combined = pd.concat([df_GPS_lon1,df_GPS_lon2])
#save combined dataset as new file:
df_GPS_lon_combined.to_csv("C:\Temp\D7015B\GPS.longitude.csv",index=None)
#free memory:
del df_GPS_lon1
del df_GPS_lon2
del df_GPS_lon_combined

#GPS Speed:
df_GPS_speed1 = pd.read_csv("C:\Temp\D7015B\Data2\\2024-12-10 10-00-00 (1)\\GPS.speed.csv", header=None)
df_GPS_speed2 = pd.read_csv("C:\Temp\D7015B\Data2\\2024-12-10 12-00-00 (1)\\GPS.speed.csv", header=None)
df_GPS_speed_combined = pd.concat([df_GPS_speed1,df_GPS_speed2])
#save combined dataset as new file:
df_GPS_speed_combined.to_csv("C:\Temp\D7015B\GPS.speed.csv",index=None)
#free memory:
del df_GPS_speed1
del df_GPS_speed2
del df_GPS_speed_combined

#Vibration 1:
df_VIB1_1 = pd.read_csv("C:\Temp\D7015B\Data2\\2024-12-10 10-00-00 (1)\\CH1_ACCEL1Z1.csv", header=None)
df_VIB1_2 = pd.read_csv("C:\Temp\D7015B\Data2\\2024-12-10 12-00-00 (1)\\CH1_ACCEL1Z1.csv", header=None)
df_VIB1_combined = pd.concat([df_VIB1_1,df_VIB1_2])
#save combined dataset as new file:
df_VIB1_combined.to_csv("C:\Temp\D7015B\CH1_ACCEL1Z1.csv",index=None)

#free memory:
del df_VIB1_1
del df_VIB1_2
del df_VIB1_combined

#Vibration 2:
df_VIB2_1 = pd.read_csv("C:\Temp\D7015B\Data2\\2024-12-10 10-00-00 (1)\\CH2_ACCEL1Z2.csv", header=None)
df_VIB2_2 = pd.read_csv("C:\Temp\D7015B\Data2\\2024-12-10 12-00-00 (1)\\CH2_ACCEL1Z2.csv", header=None)

df_VIB2_combined = pd.concat([df_VIB2_1,df_VIB2_2])
#save combined dataset as new file:
df_VIB2_combined.to_csv("C:\Temp\D7015B\CH2_ACCEL1Z2.csv",index=None)

#free memory:
del df_VIB2_1
del df_VIB2_2
del df_VIB2_combined

'''

#Caj Rollny: define the new files:
files = {
    "latitude": "C:\Temp\D7015B\GPS.latitude.csv",
    "longitude": "C:\Temp\D7015B\GPS.longitude.csv",
    "vibration1": "C:\Temp\D7015B\CH1_ACCEL1Z1.csv",
    "vibration2": "C:\Temp\D7015B\CH2_ACCEL1Z2.csv",
    "speed": "C:\Temp\D7015B\GPS.speed.csv"
}

'''
#Caj Rollny: hard-coded the file paths above, no need for opening other files for now. 
def load_file(key):
    print("In function load_file, askopen filename")
    file_path = filedialog.askopenfilename(filetypes=[("CSV Files", "*.csv")])
    if file_path:
        files[key] = file_path
        print(f"{key.capitalize()} file loaded: {file_path}")

print("Select Latitude File")
load_file("latitude")
print("Select Longitude File")
load_file("longitude")
print("Select Vibration 1 File")
load_file("vibration1")
print("Select Vibration 2 File")
load_file("vibration2")
print("Select Speed File")
load_file("speed")
'''

# Load each CSV into a DataFrame and add a 'timestamp' using the row index.
dataframes = {}
for key, file_path in files.items():
    if file_path:
        df = pd.read_csv(file_path, header=None, names=[key])
        df['timestamp'] = df.index
        dataframes[key] = df
    else:
        print(f"{key.capitalize()} file not selected.")

print("loading data in dataframes finished.")


# ====================
# Create GPS DataFrame by merging latitude and longitude.
# ====================
if "latitude" in dataframes and "longitude" in dataframes:
    df_gps = pd.merge(dataframes["latitude"], dataframes["longitude"], on="timestamp")
    # Rename columns for consistency
    df_gps = df_gps.rename(columns={"latitude": "Latitude", "longitude": "Longitude"})
    # Add an index column for use in the interactive plot
    df_gps["PointIndex"] = df_gps.index
else:
    print("Latitude or Longitude data is missing.")
    df_gps = pd.DataFrame(columns=["Latitude", "Longitude", "PointIndex"])
#debug: save ther resulting gps dataframe to file:
#df_gps.to_csv("C:\Temp\D7015B\GPS_resulting.csv")

print("creating GPS merged dataframes finished.")

# ====================
# Merge the two vibration signals on 'timestamp'
# ====================
if "vibration1" in dataframes and "vibration2" in dataframes:
    df_vibration_merged = pd.merge(
        dataframes["vibration1"],
        dataframes["vibration2"],
        on="timestamp"
        # When the column names differ (here: vibration1 vs vibration2), suffixes are not needed.
    )
    # You may rename columns if desired; here they remain "vibration1" and "vibration2"
else:
    print("Vibration data files are missing.")
    df_vibration_merged = pd.DataFrame()

print("merging vibration dataframes on timestamp finished.")

# ====================
# Data Preprocessing and Segmentation for Vibration Data
# ====================
#Caj Rollny: after investigating the files I verified that the sampling rate should be 5000Hz for vib and 5Hz for gps
#old: dt_vibration = 0.002  # seconds per sample (e.g. 500 Hz sampling rate)
dt_vibration = 0.0002  # seconds per sample (e.g. 5000 Hz sampling rate)
segment_duration_seconds = 0.2 #Caj Rollny: we want one segment per gps data point (5Hz), the segment should be 0.2s to match that.
#old:  segment_duration_seconds = 10
segment_length = int(segment_duration_seconds / dt_vibration)
print("segment_length: ", segment_length)
if not df_vibration_merged.empty:
    num_segments = len(df_vibration_merged) // segment_length
    print("num_segments: ", num_segments)
    segments = []
    for i in range(num_segments):
        seg = df_vibration_merged.iloc[i * segment_length: (i + 1) * segment_length][["vibration1", "vibration2"]].values
        segments.append(seg)
    segments = np.array(segments)
    #print("Segmented vibration data shape:", segments.shape)
    
else:
    segments = np.array([])
    print("No vibration data available for segmentation.")

#Caj Rollny: clean the data; a lot of GPS-data must be removed, look at number of satellites 
# and the points that have less than 5 satellites, also remove corresponding vibration data.
#TODO

print("Data preprocessing and segmentation for vibration data finished.")
#print("Segments: ", segments)
# ====================
# Build the Interactive Dash App
# ====================

# Create the interactive GPS map using Plotly Express.
if not df_gps.empty:
    # Use custom_data to store the point index so that it will be available in callbacks.
    map_fig = px.scatter_mapbox(
        df_gps,
        lat="Latitude",
        lon="Longitude",
        custom_data=["PointIndex"],
        zoom=10,
        title="GPS Points with Vibration Data"
    )
    map_fig.update_layout(mapbox_style="open-street-map", height=600)
else:
    map_fig = go.Figure()
    map_fig.update_layout(title="No GPS Data Available", height=600)

# Create an initial empty vibration plot figure.
vib_empty_fig = go.Figure()
vib_empty_fig.update_layout(
    title="Vibration Signal",
    xaxis_title="Time (s)",
    yaxis_title="Acceleration"
)

print("Creation of interactive map finished.")

# Initialize Dash app.
app = dash.Dash(__name__)

app.layout = html.Div([
    html.Div([
        dcc.Graph(id="gps-map", figure=map_fig)
    ], style={'width': '48%', 'display': 'inline-block', 'vertical-align': 'top'}),
    html.Div([
        dcc.Graph(id="vibration-plot", figure=vib_empty_fig)
    ], style={'width': '48%', 'display': 'inline-block', 'vertical-align': 'top'})
])
print("Initialization of Dash app finished.")

# --------------------
# Callback to Update the Vibration Plot Based on Clicked GPS Point
# --------------------
@app.callback(
    Output('vibration-plot', 'figure'),
    [Input('gps-map', 'clickData')]
)
def update_vibration_plot(clickData):
    # If no point is selected, return the empty vibration figure.
    print("Clickdata:", clickData)

    if clickData is None:
        return vib_empty_fig

    # Retrieve the selected GPS point index from custom_data.
    #this line crashed, no 'customdata' in clickData: point_index = clickData['points'][0]['customdata'][0]
    #trying this:
    point_index = clickData['points'][0]['pointIndex']
    print("point_index: ", point_index)
    
    if segments.size == 0:
        empty_fig = go.Figure()
        empty_fig.update_layout(
            title="No Vibration Data Available",
            xaxis_title="Time (s)",
            yaxis_title="Acceleration"
        )
        return empty_fig

    # Map the GPS point to a vibration segment. If the selected index exceeds
    # the available number of segments, use the last segment as fallback.
    print("segments.shape[0] = ", segments.shape[0])
    if point_index < segments.shape[0]:
        selected_segment = segments[point_index]
    else:
        selected_segment = segments[-1]
        print("selected segment =segments[-1]")

    # Create a time axis for the selected segment.
    time_axis = np.arange(segment_length) * dt_vibration

    vib_fig = go.Figure()
    vib_fig.add_trace(go.Scatter(
        x=time_axis,
        y=selected_segment[:, 0],
        mode='lines',
        name='Vibration Channel 1'
    ))
    vib_fig.add_trace(go.Scatter(
        x=time_axis,
        y=selected_segment[:, 1],
        mode='lines',
        name='Vibration Channel 2'
    ))
    vib_fig.update_layout(
        title=f"Vibration Signal for GPS Point {point_index}",
        xaxis_title="Time (s)",
        yaxis_title="Acceleration"
    )
    return vib_fig

# ====================
# Run the Dash App
# ====================
if __name__ == "__main__":
    app.run_server(debug=True, port=8060)
```
